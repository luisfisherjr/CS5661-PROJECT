```python

randomForest_obj = RandomForestClassifier(n_estimators = 19, bootstrap = True, random_state=3)
my_AdaBoost_obj = AdaBoostClassifier(randomForest_obj, n_estimators = 30, learning_rate=0.2,random_state=3)
my_AdaBoost_obj.fit(X_train, y_train)
X_predict= my_AdaBoost_obj.predict(X_test)

score = accuracy_score(y_test, X_predict)
print("Adaboost Score: ", score)

```
